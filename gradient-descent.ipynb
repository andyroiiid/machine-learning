```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_train = np.array([1, 2])
y_train = np.array([300, 500])
```

```python
def model(x, w, b):
    return w * x + b
```

```python
def cost(x, y, w, b):
    m = x.shape[0]
    f = model(x, w, b)
    cost = np.sum((f - y) ** 2) / (2 * m)
    return cost
```

```python
def plot(x, y, w, b):
    predicted = model(x, w, b)
    plt.scatter(x, y, label="Train", c="r")
    plt.plot(x, predicted, label="Predicted", c="b")
    plt.title(f"w = {w}, b = {b}, cost = {cost(x, y, w, b)}")
    plt.show()
```

```python
plot(x_train, y_train, 100, 100)
```

```python
plot(x_train, y_train, 200, 100)
```

```python
def gradient(x, y, w, b):
    m = x.shape[0]
    f = model(x, w, b)
    dj_dw = np.sum((f - y) * x) / m
    dj_db = np.sum(f - y) / m
    return dj_dw, dj_db
```

```python
def gradient_descent(x, y, w, b, alpha, iterations):
    for i in range(iterations):
        if i % 1000 == 0:
            print(f"iteration #{i}, w = {w}, b = {b}, cost = {cost(x, y, w, b)}, gradient = {gradient(x, y, w, b)}")
        dj_dw, dj_db = gradient(x, y, w, b)
        w = w - alpha * dj_dw
        b = b - alpha * dj_db
    print(f"iteration #{iterations}, w = {w}, b = {b}, cost = {cost(x, y, w, b)}, gradient = {gradient(x, y, w, b)}")
    return w, b
```

```python
w, b = gradient_descent(x_train, y_train, 100, 100, 0.01, 10000)
plot(x_train, y_train, w, b)
```
